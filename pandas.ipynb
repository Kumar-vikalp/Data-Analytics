### Forward filing and Backward filling

```python
# forward filling 
# the nan value will be updated with the data it has forward of it 

df4.windspeed.ffill()

# here in this case 
# 20.0
# 33.0
# NaN

# so all the nan was updated with 33.0

# why we use this forward fill?
# its a subsiqencial data 
# so the yesterdays temprature can be todays temprature
# so its a good approch in those cases
```

```python
# Backward filling 
df4.weather.bfill()

# here in this case 
# 7    sunny
# 8    NaN
# 9    sunny


# so all the nan was updated with 33.0
```
